```python
import csv
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("https://raw.githubusercontent.com/chandanverma07/DataSets/master/Car_sales.csv")



df.to_csv("raw_car_sales.csv", index=False)
df_dropped = df.dropna()

plt.figure(figsize=(12,6))
plt.bar(df["Manufacturer"], df["Sales in thousands"], color='red')
plt.title("Total Car Sales by Manufacturer")
plt.xlabel("Manufacturer")
plt.ylabel("Sales in thousands")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
